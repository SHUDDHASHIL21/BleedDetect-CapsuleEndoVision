## Augmentation for Bleeding cell images

```python
#input directory 
input_directory = '/content/drive/MyDrive/BLEEDGEN/data/bleeding/Images/'
input_directory_bw = '/content/drive/MyDrive/BLEEDGEN/data/bleeding/Annotations/'

#output directories for rotated images
output_directory = '/content/drive/MyDrive/BLEEDGEN/data/'


import cv2
import os
import pybboxes as pbx

# List the files in the input directory
input_files = os.listdir(input_directory)

angles = [90, 180, 270]
for angle in angles:
    for filename in input_files:
        if filename.endswith(".png"):
            print(angle,filename)
            img = cv2.imread(os.path.join(input_directory, filename))
            img_bw = cv2.imread(os.path.join(input_directory_bw, "ann- ("+filename.split('img- (')[1]))
            # print(os.path.join(input_directory_bw, "ann- ("+filename.split('img- (')[1]))
            if angle == 90:
              rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # Rotate by the specified angle
              rotated_img_bw = cv2.rotate(img_bw, cv2.ROTATE_90_CLOCKWISE)
            elif angle == 180 :
              rotated_img = cv2.rotate(img, cv2.ROTATE_180) # Rotate by the specified angle
              rotated_img_bw = cv2.rotate(img_bw, cv2.ROTATE_180)
            elif angle == 270 :
              rotated_img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) # Rotate by the specified angle
              rotated_img_bw = cv2.rotate(img_bw, cv2.ROTATE_90_COUNTERCLOCKWISE)
            output_filename = os.path.splitext(filename)[0] + f"_rotated_{angle}.png"
            cv2.imwrite(output_directory+'rot_image_'+str(angle)+'/'+output_filename, rotated_img)
            xmax= -1
            xmin= rotated_img_bw.shape[0]
            ymax= -1
            ymin= rotated_img_bw.shape[0]
            for i in range(rotated_img_bw.shape[0]):
              for j in range(rotated_img_bw.shape[1]):
                if rotated_img_bw[i][j][0] == 255:
                  if i>ymax:
                    ymax = i
                  if i<ymin:
                    ymin = i
                  if j>xmax:
                    xmax = j
                  if j<xmin:
                    xmin = j
            voc_bbox = (xmin , ymin , xmax , ymax)
            W =  rotated_img_bw.shape[1]
            H = rotated_img_bw.shape[0]
            conv = pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(W,H))
            output_filename = os.path.splitext(filename)[0] + f"_rotated_{angle}.txt"
            p = open(output_directory+'rot_label_'+str(angle)+'/'+ output_filename , 'w')
            p.write('0 '+str(conv[0])+' '+str(conv[1])+' '+str(conv[2])+' '+str(conv[3]))
            p.close()
    #     break
    # break
```
